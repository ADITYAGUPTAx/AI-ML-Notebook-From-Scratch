```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    n=len(x)
    learning_rate=0.08  #step size
    iterations=12000 #no of steps needed
    for i in range(iterations):
        y_predicted=m_curr*x +b_curr
        cost=1/n*sum((y-y_predicted)**2)
        md =(-2/n)*sum(x*(y-y_predicted))   #m deriavtive
        bd =(-2/n)*sum((y-y_predicted))   #b derivative

        m_curr =m_curr - learning_rate*md
        b_curr=b_curr -learning_rate*bd
        print("m {}, b {},     cost:{}, iteration {}".format(m_curr, b_curr,cost, i))
        



x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```

```python
def gradient_descent(x,y):
    m=0
    b=0
    LR=0.00022
    iterations=1000
    n=len(x)
    for i in range(iterations):
        y_predicted=m*x+b
        md=(-2/n)*sum(x*(y-y_predicted))
        bd=(-2/n)*sum((y-y_predicted))
        cost=(1/n)*sum((y-y_predicted)**2)
        m=m-LR*md
        b=b-LR*bd
        print(f"m: {m},b:{b},cost:{cost},iteration:{i}, y_predicted:{y_predicted}, y:{y}")

x=np.array(df.math)
y=np.array(df.cs)
gradient_descent(x,y)
```
